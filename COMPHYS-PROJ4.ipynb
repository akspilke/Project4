```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font',
**
{'family':'serif'})
import pylab
from numpy import *
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import scipy.interpolate
from scipy import stats
%matplotlib inline

E, M, t = np.loadtxt('N20T24steps1e4ORDERED.txt', usecols=(0,1,2), unpack=True) #N2T1e4#N20T24steps1e4.txt

plt.figure(1)
plt.plot(t, E, label='T=2.4')
plt.title("Energy variation with number of Cycles", size=15)
plt.ylabel("Energy", size=15)
plt.xlabel("Timesteps (Monte Carlo Cycles)", size=15)
#plt.ylim(-850, -350)
plt.xlim(0,500)
plt.grid()
plt.legend()

plt.figure(2)
plt.plot(t, M, label='T=2.4')
plt.title("Magnetic Moment variation with number of Cycles", size=15)
plt.ylabel("Magnetic Moment", size=15)
plt.xlabel("Timesteps (Monte Carlo Cycles)", size=15)
#plt.ylim(0, 450)
#plt.ylim(-100, 400)
plt.xlim(0,500)
plt.grid()
plt.legend()

pp=np.where(t>100)
Prob_E = np.where(E==-620.)

print np.size(Prob_E)

N = 20
averageEnergies = []
for i in range(N, len(E), 1):
    avgE = 0
    for j in range(N):
        avgE += E[i-j]
    avgE /= N
    averageEnergies.append(avgE)
    
plt.figure(3)
plt.plot(averageEnergies, label='T=2.4')
plt.ylim(-800, 800)
plt.title("Smoothed Energy vs. Cycles", size=15)
plt.ylabel("Energy", size=15)
plt.xlabel("Timesteps (Monte Carlo Cycles)", size=15)
plt.legend()
#plt.xlim(0,2000)
plt.grid()

averageM = []
for i in range(N, len(M), 1):
    avgM = 0
    for j in range(N):
        avgM += M[i-j]
    avgM /= N
    averageM.append(avgM)
plt.figure(4)
plt.plot(averageM, label='T=2.4')
plt.ylim(-800, 800)
plt.title("Smoothed Magnetic Moment vs. Cycles", size=15)
plt.ylabel("Magnetic Moment", size=15)
plt.xlabel("Timesteps (Monte Carlo Cycles)", size=15)
plt.legend()
#plt.xlim(0,2000)
plt.grid()
```
